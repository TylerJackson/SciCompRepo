```python
def omega(x,t):
    product=1.0;
    for i in range(11):
       product=product*(t-x[i]);
    return product;

t=linspace(-5.0,5.0,201)
omegaSol=zeros((201));
for i in range(201):
    omegaSol[i]=omega(chebyNodes,t[i])

plot(t,omegaSol)
```

```python
regularnodes=linspace(-5.0,5.0,11)
omegaSol=zeros((201));
for i in range(201):
    omegaSol[i]=omega(regularnodes,t[i])

plot(t,omegaSol)
```
